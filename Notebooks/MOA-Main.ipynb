```python
import pandas as pd
```

```python
def read_data():
    train = pd.read_csv('../data/train_features.csv')
    train_labels = pd.read_csv('../data/train_targets_scored.csv')
    test = pd.read_csv('../data/test_features.csv')
    train = train.merge(train_labels)
    
    train = train.set_index('sig_id')
    test = test.set_index('sig_id')
    
    del train_labels
    
    return train, test
```

```python
train, test = read_data()
```
